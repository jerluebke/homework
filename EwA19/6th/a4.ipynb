```python
import numpy as np
from scipy.integrate import quad
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['figure.dpi'] = 300
```

## Teil a

```python
def gaussian(x):
    a = .5
    return a*np.exp(-(a*x)**2)
```

```python
np.sqrt(np.pi)/2
```

```python
quad(gaussian, 0, np.inf)[0]
```

```python
# error
abs(quad(gaussian, 0, np.inf)[0] - np.sqrt(np.pi)/2)
```

## Teil b

```python
def my_gammainc(x, s):
    return quad(lambda t: t**(s-1)*np.exp(-t), 0, x)[0]
```

```python
s = 3
my_gammainc(4, s)
```

## Teil c

```python
x = np.linspace(0, 3*s, 100)
y = np.array([my_gammainc(z, s) for z in x])
plt.plot(x, y)
plt.gca().set_title(label='$\gamma(x, s=3) = \int_{0}^{x} t^{s-1}\,e^{-t} \mathrm{d}x$')
```

## Teil d

```python
from scipy.special import gamma, gammainc
```

```python
y_scipy = gamma(3) * gammainc(3, x)
plt.plot(x, y, label='my gamma function')
plt.plot(x, y_scipy, label='scipy.special.gamminc')
plt.gca().set_title('Vergleich: Incomplete $\Gamma$ Function Implementierungen')
plt.legend()
```

```python
# mean error
np.mean(np.abs(y-y_scipy))
```
